```python
import bisect

#Nice
class SpellChecker:
    def __init__(self, dictionary):
        # Sort the dictionary in lexicographic order and store it as a list
        self.dictionary = sorted(dictionary)

    def find_nearest_words(self, word):
        # Use binary search to find the closest word in the dictionary
        index = bisect.bisect_left(self.dictionary, word)

        # Get the nearest 4 words (2 on each side of the index)
        nearest_words = self.dictionary[max(0, index - 2):index] + self.dictionary[index:min(len(self.dictionary), index + 2)]

        return nearest_words

    def add_to_dictionary(self, word):
        # Use binary search to find the position to insert the new word in lexicographic order
        index = bisect.bisect_left(self.dictionary, word)

        # Inserting  word into the dictionary at the correct position
        self.dictionary.insert(index, word)

    def print_dictionary(self):
        # Print the dictionary in lexicographic order
        print("\n".join(self.dictionary))




        #ensuring that we read English letters only
def read_text_file(file_path, encoding='cp1252'):
    result_dict = {}
    with open(file_path, 'r', encoding=encoding, errors='replace') as file:
        for line in file:
            # Remove non-ASCII characters
            english_letters = ''.join(char for char in line if 65 <= ord(char) <= 90 or 97 <= ord(char) <= 122)
            key = english_letters.strip()
            if key:
                result_dict[key] = 1
    return result_dict


file_path = '/kaggle/working/dictionary.txt'

#Using cp1252 encoding as it helps to handle special characters
dictionary_data = read_text_file(file_path,'cp1252')
# Create the spell checker with the given dictionary data
spell_checker = SpellChecker(dictionary_data)

# Test the spell checker
input_word = 'abacu'
print(f"Nearest 4 words to '{input_word}': {spell_checker.find_nearest_words(input_word)}")

# Adding of  a new word to the dictionary
new_word = 'abacusical'
spell_checker.add_to_dictionary(new_word)


```
